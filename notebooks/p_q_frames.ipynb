```python
def calculate_intervals(data, percentage):

    lower_percentile = (100 - percentage) / 2

    upper_percentile = 100 - lower_percentile

    interval_mean = (

        data["mean"].quantile(lower_percentile / 100),

        data["mean"].quantile(upper_percentile / 100)

    )

    interval_std = (

        data["std"].quantile(lower_percentile / 100),

        data["std"].quantile(upper_percentile / 100)

    )

    return interval_mean, interval_std
```

```python
charge_threshold = 0

# Identify events where the maximum charge exceeds the threshold
max_charge_per_event = df_q.groupby('Event')['charge'].max()
filtered_events = max_charge_per_event[max_charge_per_event > charge_threshold].index

# Filter the dataset to include only these events
df_filtered = df_q[df_q['Event'].isin(filtered_events)]
```

```python
# Part 2: Heat maps of charge per time for one 'Event'
def plot_heatmap_single_event(data, event_id, time_bins=100, charge_bins=40):
    single_event_data = data[data['Event'] == event_id]
    if single_event_data.empty:
        print(f"No data available for Event {event_id}.")
        return

    # Create a 2D histogram (heatmap) for log scale
    heatmap_log, xedges_log, yedges_log = np.histogram2d(
        single_event_data['time'], 
        single_event_data['charge'], 
        bins=[time_bins, charge_bins], 
        range=[[single_event_data['time'].min(), single_event_data['time'].max()], [0, 35]]
    )

    # Apply log scale to the heatmap counts
    heatmap_log = np.log10(heatmap_log + 1)  # Add 1 to avoid log(0)

    # Plot the heatmap with log scale
    plt.figure(figsize=(10, 6))
    plt.imshow(
        heatmap_log.T, origin='lower', aspect='auto',
        extent=[xedges_log[0], xedges_log[-1], yedges_log[0], yedges_log[-1]]
    )
    plt.colorbar(label='Log(Count + 1)')
    plt.xlabel('Time')
    plt.show()

    # Create a 2D histogram (heatmap) for linear scale
    heatmap_linear, xedges_linear, yedges_linear = np.histogram2d(
        single_event_data['time'], 
        single_event_data['charge'], 
        bins=[time_bins, charge_bins], 
        range=[[single_event_data['time'].min(), single_event_data['time'].max()], [0, 5]]
    )

    # Plot the heatmap with linear scale
    plt.figure(figsize=(10, 6))
    plt.imshow(
        heatmap_linear.T, origin='lower', aspect='auto',
        extent=[xedges_linear[0], xedges_linear[-1], yedges_linear[0], yedges_linear[-1]]
    )
    plt.colorbar(label='Count')
    plt.xlabel('Time')
    plt.ylabel('Charge')
    plt.title(f'Charge vs Time Heatmap (Linear Scale, 0-5) for Event {event_id}')
    plt.savefig('plots/charge_time_1_lin.png')
    plt.show()
    
# Select an example event to plot (first event in filtered dataset)
if not filtered_events.empty:
    example_event_id = filtered_events[4000]
    plot_heatmap_single_event(df_filtered, example_event_id)
    
else:
    print("No events meet the charge threshold.")
```

```python
def plot_histogram_counts_per_time(data, event_id, time_bins=100):
    single_event_data = data[data['Event'] == event_id]
    if single_event_data.empty:
        print(f"No data available for Event {event_id}.")
        return

    # Calculate mean and standard deviation of the charge
    charge_mean = single_event_data['charge'].mean()
    charge_std = single_event_data['charge'].std()

    # Define charge intervals based on mean and standard deviation
    charge_intervals = [
        (charge_mean - charge_std, charge_mean + charge_std),
        (charge_mean + charge_std, charge_mean + 2 * charge_std),
        (charge_mean + 2 * charge_std, charge_mean + 3 * charge_std)
    ]

    # Initialize the histogram data
    plt.figure(figsize=(10, 6))

    # Loop over charge intervals to calculate and add counts to the histogram
    for i, (charge_min, charge_max) in enumerate(charge_intervals):
        counts, time_edges = np.histogram(
            single_event_data[(single_event_data['charge'] >= charge_min) & (single_event_data['charge'] < charge_max)]['time'],
            bins=np.linspace(single_event_data['time'].min(), single_event_data['time'].max(), time_bins + 1)
        )
        time_centers = (time_edges[:-1] + time_edges[1:]) / 2
        plt.hist(
            time_centers, bins=len(time_centers), weights=counts, histtype='step',
            label=f'Mean ± {i + 1}σ', linewidth=1.5
        )

    # Finalize the histogram plot
    plt.xlabel('Time')
    plt.ylabel('Counts')
    plt.yscale('log')
    plt.title(f'Counts per Time Histogram for Event {event_id} by Charge Intervals')
    plt.legend()
    plt.show()

# Select an example event to plot (first event in filtered dataset)
if not filtered_events.empty:
    example_event_id = filtered_events[200]
    plot_histogram_counts_per_time(df_filtered, example_event_id)
else:
    print("No events meet the charge threshold.")


```
